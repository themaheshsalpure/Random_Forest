```python
df = pd.read_csv('D:\\1-Data Science\\6 - Machine Learning\\Random Forest\\Data\\Diabetes.csv')
```

```python
df.rename(columns=lambda x: x.strip(), inplace=True)
```

```python
df.columns
```

```python
target = df['Class variable']
input = df.drop('Class variable', axis = 'columns')
```

```python
input
```

```python
def norm(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
input_n = norm(input)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# md = MinMaxScaler()
# input_n = md.fit_transform(input)
```

```python
input_n
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
X_train, X_test, y_train, y_test = train_test_split(input_n, target, test_size = 0.2)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
pred = model.predict(X_test)
```

```python
pred
```

```python
cm = confusion_matrix(pred,y_test)
```

```python
cm
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True)
plt.xlabel("Prediction")
plt.ylabel('Truth')
```
